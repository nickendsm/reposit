```python
import sys  # sys нужен для передачи argv в QApplication
from PyQt5 import QtCore, QtGui, QtWidgets
import matplotlib
matplotlib.use('Qt5Agg')
from PyQt5.QtCore import QTimer
import random
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg
from matplotlib.figure import Figure
import matplotlib.pyplot as plt

class MplCanvas(FigureCanvasQTAgg):
    def __init__(self, parent=None, width=5, height=4, dpi=100):
        fig = Figure(figsize=(width, height), dpi=dpi)
        self.axes = fig.add_subplot(111)
#         self.axes1 = fig.add_subplot(132)
#         self.axes2 = fig.add_subplot(133)
        super(MplCanvas, self).__init__(fig)

class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(800, 500)
        self.sc = MplCanvas(self, width=4, height=4, dpi=100)
#         self.sc.axes.plot([0,1,2,3,4], [10,1,20,3,40])
        self.setCentralWidget(self.sc)
#         self.centralwidget = QtWidgets.QWidget(MainWindow)
#         self.centralwidget.setObjectName("centralwidget")
        self.pushButton = QtWidgets.QPushButton(self.sc)
        self.pushButton.setGeometry(QtCore.QRect(650, 460, 80, 30))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.TimerStop)
        self.pushButton_2 = QtWidgets.QPushButton(self.sc)
        self.pushButton_2.setGeometry(QtCore.QRect(500, 460, 80, 30))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.TimerStart)
#         MainWindow.setCentralWidget(self.centralwidget)
#         self.menubar = QtWidgets.QMenuBar(MainWindow)
#         self.menubar.setGeometry(QtCore.QRect(0, 0, 342, 18))
#         self.menubar.setObjectName("menubar")
#         MainWindow.setMenuBar(self.menubar)
#         self.statusbar = QtWidgets.QStatusBar(MainWindow)
#         self.statusbar.setObjectName("statusbar")
#         MainWindow.setStatusBar(self.statusbar)
        # канвасы
        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)
        
        self.show()
        
    def TimerStart(self):
        self.timer = QtCore.QTimer()
        self.timer.setInterval(100)
        self.timer.timeout.connect(self.update_plot)
        self.timer.start()
    
    def TimerStop(self):
        self.timer.stop()
        self.sc.axes.cla()
        self.sc.draw()
        
    def update_plot(self):
        self.generate_random()
        self.sc.axes.plot(self.x,self.y)
        self.sc.draw()
        
    def clearCanv(self):
        self.sc.axes.cla()
        self.sc.draw()

    def generate_random(self):
        self.x = [int(random.randint(0, 1)) for i in range(10)]
        self.y = [int(random.randint(0, 1)) for i in range(10)]
#         print(self.x,self.y)
        
    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.pushButton.setText(_translate("MainWindow", "Стереть"))
        self.pushButton_2.setText(_translate("MainWindow", "Построить"))


class ExampleApp(QtWidgets.QMainWindow, Ui_MainWindow):
    def __init__(self):
        super().__init__()
        self.setupUi(self) 
        
def main():
    app = QtWidgets.QApplication(sys.argv)  # Новый экземпляр QApplication
    window = ExampleApp()  # Создаём объект класса ExampleApp
    window.show()  # Показываем окно
    app.exec_()  # и запускаем приложение
main()
```
