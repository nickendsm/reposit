* `CTRL` + `/` - установить/снять символ комментария у выделенных строк.

## Базовые типы данных Python 

Все данные, которые обрабатываются в программах на Python, называются *объектами*.  

**(!)** Термин *объект* в Python не означает экземляр класса, как например, в C++. 

"Из коробки" в Python доступно большое количество встроенных типов данных:
* числовой тип (целый `int`, вещественный `float`, комплексный `complex`),
* логический тип `bool`,
* строковый `str`,
* кортеж `tuple`,
* список `list`,
* словарь `dict`,
* функция `function`,
* класс `type`,
* модуль `module`,
* ...

Имеется также специальный тип `NoneType` и единственный объект этого типа `None` - пустой объект.

**(!)** Python является языком с динамической типизацией, т.е. тип переменной определяется во время исполнения программы.

**(!)** Все объекты можно условно разделить на *атомарные* и *контейнерные* (*составные*), последние, в свою очередь, бывают *изменяемыми* и *неизменяемыми*. К составным типам данных относятся *строки*, *кортежи*, *списки*, *словари*, *множества*.

**(!!)** При выполнении операции присваивания происходит копирование ссылки на объект. Это является трудным моментом для начинающих программистов Python, поэтому иногда встречаются рекомендации вместо термина "присваивание значения переменной" употреблять термин "связывание значения с именем переменной".

Функция `type(object)` позволяет узнать тип объекта `object`.  
Функция `isinstance(object, type)` возвращает `True`, если объект `object` имеет тип `type`.  
Функция `id(object)` позволяет узнать уникальный идентификатор объекта `object`.  
Функция `len` позволяет узнать длину составного объекта `object`.

Преобразование в требуемый тип выполняется одноимённой функцией, например, функция `int()` преобразует свой аргумент в целое число.

### Логические и физические строки

Программа на Python состоит из *логических строк*.  
Логические строки могут состоять из одной или нескольких *физических строк*.

Физическая строка заканчивается символом перевода строки.  
Физические строки могут быть объединены в одну логическую строку с помощью символа `\`.

Текст, следующий за символом решетки `#` является *комментарием*.  
Символ решётки нельзя использовать на строке, которая заканчивается символом `\`.

Текст, заключенный в круглые `()`, квадратные `[]` и фигурные скобки `{}`,  
можно располагать на нескольких строках, не используя символ `\`.

````python
# Правильно:
x = 1 + 2 + 3 + 4

print(x)
````

````python
# Неправильно:
#
# x = 1 + 2 + 
#     3 + 4
# print(x)

# Ошибка:
# File "<ipython-input-4-4c19256484cd>", line 3
#     x = 1 + 2 +
#                 ^
# SyntaxError: invalid syntax
````

````python
# Правильно:
x = 1 + 2 + \
    3 + 4

print(x)
````

````python
# Правильно:
x = (1 + 2 + 
     3 + 4)

print(x)
````

````python
# Неправильно:
# x = 1 + 2 + \ # здесь нельзя использовать комментарий
#     3 + 4
# print(x)

# Ошибка:
# File "<ipython-input-8-d0cdedad5af8>", line 2
#     x = 1 + 2 + \ # здесь нельзя использовать комментарий
#                                                           ^
# SyntaxError: unexpected character after line continuation character
````

````python
x = (1 + 2 + # здесь можно использовать комментарий
     3 + 4)

print(x)
````

### Отступы строк

В Python нет какого-либо специального символа или зарезервированных слов для оформления границ блоков программы. Оформление блоков осущетсвляется с помощью *отступов*, для установки которых можно использовать символы пробела или табуляцию. Строки, имеющие одинаковый отступ, принадлежат одному *блоку*.

````python
# Правильно:
x = 1
if (x > 0) & (x > -5):
    print('x is positive')
else:
    print('x is negative')
````

### Оператор удаления объекта del

Оператор удаления `del` производит удаление привязки имени объекта к его содержимому.

**(!)** Фактического удаления данных может не произойти, если имеется другая переменная, ссылающаяся на эти же данные.

**(!)** Оператор `del` также может быть использован для удаления атрибутов у объектов и элементов у изменяемых составных объектов.

````python
x = 1

del(x)
````

````python
x = [1, 2]
y = x

del(x)    # фактического удаления списка не происходит, т.к. имеется ссылка на него (переменная y)
print(y)
del(y[0]) # удаление первого элемента списка y
print(y)
````

### Арифметические операторы

В Python существуют следующие арифметические операции:
* `+` - сложение,
* `-` - вычитание,
* `*` - умножение,
* `/` - деление,
* `//` - целочисленное деление,
* `%` - остаток от деления,
* `**` - возведение в степень.

**(!)** Операция возведения в степень имеет самый высокий приоритет, кроме ситуации, когда справа стоит унарный или битовый оператор.

**(!)** При выполнении целочисленного деления результат всегда округляется в меньшую сторону.

**(!)** Операторы `+` и `*` при работе с составными типами данных *строка*, *кортеж* и *список* имеют следующую семантику:  
* `x + y` - объединение объектов `x` и `y`,
* `n * x` и `x * n` - объединение `n` копий объекта `x`.

````python
x = 5
y = 2

print(x + y)
print(x * y)
print(x / y)
print(x // y)
print(x % y)
print(x ** y)
````

````python
x = y = 2

print(-x ** -y)
````

````python
print(+5 // 2)
print(-5 // 2)
print(2*(-5 // 2) + 5%2)
````

````python
x = y = '1'

print(x + y)
print(2 * x)
````

### Логические операторы

В Python существуют следующие логические операторы:  
* `and` - логическое `AND`,
* `or` - логическое `OR`,
* `not` - логическое `NOT`.

**(!)** Операторы `and` и `or` всегда возвращают значение одного из своих операндов, причём, операнд справа может не вычисляться, если в этом нет необходимости.

````python
x = 1
y = 2
z = 3

print((x < y) and (y > z))
print((x < y) or  (y > z))
print(not(x < y))
````

### Битовые операторы

В Python существуют следующие битовые операции:
* `&` - побитовое AND,
* `|` - побитовое OR,
* `^` - побитовое XOR,
* `>>` - битовый сдвиг вправо,
* `<<` - побитовый сдвиг влево.

````python
x = 4         # в двоичной системе 00000100
y = 5         # в двоичной системе 00000101

print(x & y)  # побитовое AND = 00000100
print(x | y)  # побитовое OR  = 00000101
print(x ^ y)  # побитовое XOR = 00000001
print(x >> 1) # побитовый сдвиг вправо = 00000010  / эквивалентно операции деления на 2
print(x << 1) # побитовый сдвиг влево  = 00001000 / эквивалентно операции умножения на 2
````

### Операторы сравнения

В Python существуют следующие операторы сравнения:
* `==` - равенство,
* `!=` - неравенство,
* `<` -  меньше,
* `<=` - меньше или равно,
* `>` - больше,
* `>=` - больше или равно.

**(!)** Допускается запись `x < y < z`, которая интерпретируется как `(x < y) and (y < z)`. При этом `y` вычисляется только один раз и сравнение `(y < z)` не производится, если `(x < y)` ложно.

````python
x = 1
y = 2

print(x == y)
print(x != y)
print(x <  y)
print(x <= y)
print(x >  y)
print(x >= y)
````

### Операторы тождественности

Операторы тождественности позволяют сравнить размещение объектов в памяти:

* `is` - проверяет, что объект слева совпадает с объектом справа,
* `is not` - проверяет, что объект слева не совпадает с объектом справа,

````python
x = 1
y = x
z = x/y

print(x is y)
print(x is z)
print(x == z)
````

````python
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(y is x)     # объект y тождественен объекту x
print(z is x)     # объект z не тождествене объекту x
print(x == z)     # но при этом равен ему
````

### Составная форма оператора присваивания

Составная форма оператора присванивания имеет вид `#=`, где решётка один из символов (`+`, `-`, `*`, `/`, `%`, `//`, `**`, `&`, `|`, `^`, `>>`, `<<`). В результате выполнения оператора левому операнду присваивается значение операции `#` между левым и правым операндами.

**(!)** При работе с составными объектами строка, кортеж и список, составные формы оператора присваивания `+=` и `*=` имеют следующую семантику:  
* `x += y` эквивалентно `x = x + y`;
* `x *= n` эквивалентно `x = n * x`.

**(!)** Причём, если `x` является изменяемым объектом, то операция будет производиться без создания нового объекта.

````python
x = 1
y = 1

x += y   # эквивалентно записи x = x + y
print(x)
````

````python
x = '1'
y = '2'

print(id(x))
x += y       # x - неизменяемый составной объект, поэтому создаётся новый объект для хранения результата
print(x)
print(id(x))
````

````python
x = [1]
y = [2]

print(id(x))
x += y       # x - изменяемый составной объект, создаётся новый объект для хранения результата не создаётся 
print(x)
print(id(x))
````

## Составные операторы

Операторы в Python можно разделить на две группы: *простые* и *составные*.  

Составные операторы содержат в себе другие операторы и управляют их выполнением.  
К составным операторам относятся операторы 
* `if` - оператор ветвления;
* `while` - оператор цикла для выполнения операций, пока выполняется заданное условие;
* `for` - оператор цикла для перебора элементов;
* `try` - оператор обработки исключительных ситуаций.

В отношении записи составных операторов действую следующие правила:
* составной оператор состоит из одной или нескольких *ветвей* (частей);
* каждая ветвь состоит из *заголовка* и *тела*;
* заголовки ветвей одного составного оператора должны иметь одинаковый уровень отступа;
* заголовок ветви начинается с ключевого слова и заканчивается двоеточием;
* тело представляет собой один или несколько простых операторов, расположенных в следующих за заголовком строках и имеющих одинаковый отступ;

**(!)** Допускается расположение нескольких *простых операторов*, разделённых символом `;`, на одной строке с заголовком ветви.

### Оператор ветвления if

Оператор `if` предназначен для выполнения различных действий в зависимости от условий. Синтаксис оператора имеет вид:

```python
if Условие 1:
    Блок операторов 1
elif Условие 2:
    Блок операторов 2
elif Условие 3:
    Блок операторов 3
    
elif Условие N:
    Блок операторов N
else Условие 0:
    Блок операторов 0
```

Семантика оператора `if` следующая:  
* если первое условие истинно, то выполняется первый блок операторов,
* иначе проверяется истинность второго условия;  
* если второе условие истинно, то выполнятеся второй блок операторов,  
* иначе проверяется истинность третьего условия и т.д....;
* если все условия 1 - N оказались ложными, то выполняется блок операторов 0.

**(!)** Обязательной частью оператора является только ветвь `if`, остальные ветви могут отсутствовать. 

**(!)** При использовании оператора `if` может оказаться полезной *пустая инструкция* `pass`, которая может применять в ситуациях, когда синтаксис требует наличие инструкции, но ничего выполнять не требуется.

### Оператор цикла while

Оператор `while` предназначен для многократного выполнения действий, пока является истинным заданное условие. Синтаксис оператора имеет вид:

```python
while Условие:
    Блок операторов 1 # тело цикла
else:
    Блок операторов 0
```

Семантика оператора `while` следующая:  
* если условие истинно, то выполняется блок операторов 1;
* иначе выполняется блок операторов 0 (при наличии).

**(!)** Обязательной частью оператора является только первая ветвь `while`, ветвь `else` может отсутствовать.

**(!)** Если в теле цикла (в блоке операторов 1) происходит выполнение оператора `continue`, то происходит немедленный переход на начало цикла (остаток тела цикла игнорируется).

**(!)** Если в блоке операторов 0 происходит выполнение оператора `break` или оператора `continue`, то считается, что эти операторы относятся к внешнему циклу.

### Оператор цикла for

Оператор `for` предназначен для перебора элементов последовательности. Синтаксис оператора имеет вид:

```python
for var in sequence:
    Блок операторов 1 # тело цикла
else:
    Блок операторов 0
```
Здесь `var` - *переменная цикла*, а `sequence` - это выражение, которое вычисляется один раз в начале цикла и результатом которого должен быть объект, поддерживающий операцию извлечения элемента по индексу и генерирующий исключение `IndexError` при выходе индекса за допустимые пределы последовательности. Простейшим примером такого объекта является список.

Семантика оператора `for` следующая:
* для каждого элемента `item` последовательности, начиная с первого, выполняется присваивание `var = item`, после чего выполняется блок операторов 1;
* если последовательность исчерпана, выполняется блок операторов 0 (при наличии).

**(!)** Обязательной частью оператора является только первая ветвь `for`, ветвь `else` может отсутствовать. 

**(!)** Если в теле цикла (в блоке операторов 1) происходит выполнение оператора `break`, то происходит немедленный выход их цикла, при этом блок операторов 0 не выполняется.

**(!)** Если в теле цикла (в блоке операторов 1) происходит выполнение оператора `continue`, то происходит немедленный переход на начало цикла (остаток блока операторов 1 игнорируется).

**(!)** Если в блоке операторов 0 происходит выполнение оператора `break` или оператора `continue`, то считается, что эти операторы относятся к внешнему циклу.

**(!)** Допускается выполнять присваивание переменной `var` в теле цикла (в блоке операторов 1), это никак не повлияет на выборе следующего элемента последовательности.

**(!)** Переменная цикла `var` сохраняет своё значение после выхода из цикла, но если последовательность пустая, то инициализации переменной `var` не происходит. Без дополнительной проверки возникает риск получить ошибку `NameError`.

**(!)** Допускается изменение последовательности в теле цикла (в блоке операторов 1), но при этом следует иметь в виду, что удаление из последовательности элемента до текущего элемента приведёт к пропуску следующего элемента последовательности, а вставка нового элемента - к повторной обработке текущего элемента последовательности.

````python
# Вычисление суммы элементов последовательности x.

x = [1,2,3,4,5]
s = 0

for i in x:
    print(i)
    s += i
    
print(s)
````

### Определение функции

Объявление функции пользователя имеет вид:

```python
def ИмяФункции(Список аргументов):
    ...
    Тело функции
    ...
    return(Возвращаемое значение)
```
Здесь ключевое слово `def` является *командой*, которая предписывает интерпретатору Python создать объект - функцию - и присвоить ей указанное имя функции.

**(!!)** Ключевое слово `def` действительно является командой, которая производит связывание объекта-функции с именем переменной. Это означает, что до тех пор, пока не отработает команда `def`, функция не существует.

**(!)** Фнукция в Python - это такой же объект, как и числа, строки, и т.д. У функции, как объекта, может быть несколько имён, её можно хранить, например, в списке, и передавать в качестве аргумента в другие функции.

Команда `return` предписывает завершить выполнение функции и вернуть указанное в скобках значение в точку вызова функции. Если команда `return` отсутвует или функция завершает свою работу без этой команды, то возвращаемым значением функции будет `None`.

**(!)** Функции, которые не возвращают значения, иногда называют *процедурами*.

*Тело функции* - это фрагмент текста программы, который связывается с именем функции и который будет исполняться каждый раз при вызове функции.

*Список аргументов* - это список, состоящий из нуля или более имен аргументов функции.

**(!!)** При передаче значений аргументов в функцию используется стратегия "*вызов с разделением ресурсов*" (*call by sharing*). Для данной стратегии предполагается, что все значения в языке "*упакованы*" в объекты (*boxed*) и при вызове функция получает не сам объект, а *копию ссылки на объект*. Таким образом, сам объект не копируется, что делает данную стратегию очень похожей на стратегию "*вызов по ссылке*". Нюанс заключается в том, что присваивание аргументу в теле функции не имеет эффекта в вызывающем её контексте, т.к. происходит новое связывание имени переменной с новым значением, в то же время присваивание новых значений компонентам аргумента приводит к изменениям в вызывающем контексте.

````python
def my_fun(x):       # функция возвращает удвоенное значение своего аргумента
    result = 2*x
    return(result)
    
my_fun(1)
````

````python
def sum1(a,b=3, c=4):
    return a+b  +c

sum1(1)
````

````python
f = sum1(5)
print(f)
````

````python
def my_factorial(n):
    if n <= 0: return(1)
    return(n * my_factorial(n-1))

my_factorial(5)
````

### Сопоставление аргументов и значения по умолчанию

*Сопоставление аргументов* - это правила, в соответствии с которыми происходит сопоставление значений аргументов с их именами. Существует несколько способов сопоставления аргументов. Рассмотрим следующие из них:
* сопоставление *по позиции*,
* сопоставление *по имени*.

Сопоставление *по позиции* - это способ сопоставления значений аргументов с их именами, применяемый по умолчанию, сопоставление происводится по порядку в направлении слева направо.

Сопоставление *по имени* - сопоставление производится в соответствием с именами аргументов, порядок аргументов в этом случае не имеет значения.

**(!)** Допускается совмещение этих способов сопоставления, при условии, что вначале выполняется позиционное сопоставление, а затем по имени.

*Значение по умолчанию* - это значение, которое присваивается аргументу на этапе объявление функции и которое будет использовано при вызове функции с пропущенным значением этого аргумента.

````python
def fun(x, y, z):
    result = x*y + z
    return(result)
````

````python
# Сопоставление по позиции, x = 1, y = 2, z = 3.
fun(1, 2, 3)
````

````python
# Сопоставление по имени, x = 1, y = 2, z = 3.
fun(z = 3, y = 2, x = 1)
````

````python
# Смешанное сопоставление, x = 1, y = 2, z = 3.
fun(1, z = 3, y = 2)
````

````python
# fun(z = 3, 1, y = 2) - ошибка - сопоставление по имени предшествует позиционному сопоставлению
````

````python
def fun(x = 1, y = 0, z = 0): # значения по умолчанию аргумента x и аргумента y равно 0
    result = x*y + z
    return(result)
````

### Создание и использование модулей

Создать модуль можно в любом текстовом редакторе - достаточно написать программу на Python и сохранить её в файле с расширением `.py`.

Чтобы задействовать содержимое модуля необходимо выполнить операцию *импорта*. В простейшем случае импорт модуля осуществляется командой `import`: 

```python
import module
```  

Получить список имен объектов, определённых в модуле `module`, можно с помощью функции `dir`:
```python
dir(module)
```

**(!)** В результате импорта все инструкции, содержащиеся в файле модуля, будут выполнены, при этом в основном тексте программы становится видимым только сам импортированный модуль, а объекты, определённые в импортируемом модуле, остаются "невидимыми" и для доступа к ним необходимо указывать имя модуля.

**(!)** Инструкции, содержащиеся в модуле, исполняются *только один раз* во время его первого импорта.

**(!)** В пределах импортируемого модуля его имя хранится в переменной `__name__`.

**(!)** Поиск импортируемого модуля осуществляется в текущем каталоге, затем в каталогах списка `sys.path`, который инициализируется значением системной переменной `PYTHONPATH`. Допускается программно изменять порядок и состав списка путей поиска модулей.

**(!)** Список импортированных модулей можно получить, обратившись к словарю `sys.modules`.

**(!)** Обычно среда программирования настроена таким образом, что при импорте `.py`-модуля автоматически создаётся `.pyc`- или `.pyo`-файл (зависит от того, используется ли опция интерпретатора `-O` или `-OO`), содержащий скомпилированное содержимое `.py`-файла. Скомпилированный файл содержит время создания исходного `.py`-файла. При попытке произвести импорт модуля интерпретатор сравнивает время создания исходного файла и его скомпилированной версии и при обнаружении расхождения автоматически производит компиляцию заново. Скомпилированные версии файлов загружаются быстрее и позволяют распространять программы, написанные на Python, без включения в состав файлов с исходными текстами.

**(!)** Выполнить повторный импорт модуля без прерывания работы интерпретатора (полезно на этапе отладки приложения или в приложениях, не допускающих остановки) можно с помощью функции `reload` модуля `imp`:  
```python
from imp import reload
reload(module);
```
**(!!)** При этом следует помнить, что изменения будут "видны" только в тех частях программы, где произойдёт вызов `import` или `from`. Там, где такой вызов уже произошёл, изменения видны *не будут* и работа будет продолжена с уже существующими объектами.

**(!)** Так как импортируемые модули становятся обычными переменными, имена им следует давать в соответствие с требованиями к именам переменных, например, нельзя дать модулю имя `for`, т.к. это зарезервированное слово, попытка импорта закончится сообщением об ошибке.

**Пример**  
Требуется импортировать модуль `my_module.py` со следующим содержимым:

```python

  return(x)

print(__name__)
```

````python
import my_module # первая операция импорта
````

````python
# Использование объектов модуля.
print(my_module.my_const)
print(my_module.my_function(1))
print(my_module.__name__)
````

### Различные варианты импорта модулей

Существует две инструкции импорта модуля: `import` и `from`.

Инструкция `import` приводит к импорту содержимого модуля и созданию переменной, связанной с этим модулем. При обращении к объектам модуля необходимо указывать имя модуля с точкой перед именем объекта:

```python
import module # импорт модуля `module`
module.fun()  # вызов функции fun()
```

Инструкция `from` делает то же, что инструкция `import`, но при этом помещает имя импортируемого объекта в текущее пространство имён, что делает доступным объект для непосредственного использования:

```python
from module import fun # импорт из модуля module функции fun
fun()                  # вызов функции
```

**(!)** Вариант инструкции `from` с символом `*` импортирует содержимое указанного модуля и делает все его объекты доступными в текущем пространстве имён. Не рекомендуется использовать такой вариант инструкции `from`, т.к. это приводит к "засорению" пространства имён.

**(!)** Не допускается использовать вариант инструкции `from ... import *` внутри функций (справедливо для Python 3).

**(!)** Хотя инструкция `from` делает описание переменных менее информативным, чем инструкция `import` (имя переменной `var` менее информативно, чем имя `module.var`), тем не менее она достаточно удобна и применяется часто. Существенным недостатком этой инструкции является потенциальная возможность "разрушения" пространства имён в результате импорта одноимённых объектов из разных модулей.

Инструкции `import` и `from` допускают импорт модулей с использованием псевдонимов:

```python 
import module as m # импорт модуля module под псевдонимом "m"
m.fun()            # вызов функции fun() из модуля module

from module import fun as f # импорт из модуля module функции fun под псевдонимом "f"
f()                         # вызов функции fun из модуля module
```

**(!)** Если имя импортируемого модуля становится известно только во время выполнения программы,  
то для его импорта следует использовать команду `exec`:
```python
exec('import ' + module_name)
```

````python
from my_module import my_function
print(my_function(1))
````

````python
import my_module as my_lib
print(my_lib.my_function(1))
````

````python
from my_module import my_function as my_fun
print(my_fun(1))
````

````python
exec('import ' + 'my_module')
````

### Создание и использование пакетов

Создание пакета сводится к созданию иерархии дерева каталогов, в которых размещаются файлы модулей и файлы `__init__.py`.  
Примерная структура пакета имеет вид:  
```python
packet/
  __init__.py
  subpacket1/
    __init__.py
    module1.py
    module2.py
    ...
  subpacket2/
    __init__.py
    module1.py
    module2.py
    ...
  ...
```

Файл `__init__.py` выполняет следующие функции:  
* наличие данного файла в каталоге предписывает интерпретатору Phyton считать каталог частью пакета (в случае отсутствия файла `__init__.py` каталог и всё его содержимое будет проигнорировано, т.е. файл выполняет защитную функцию, не позволяя интерпретатору "зацепить" нежелательные файлы из каталогов, которые встретились ему в списке путей поиска модулей);  
* файл может содержать код инициализации пакета;
* файл может содержать переменную `__all__`, представляющую собой список модулей, которые необходимо загружать при выполнении команды `from ... import *`.

Импорт объектов из пакетов осуществляется также с помощью команд `import` и `from`.

**Пример**  
Имеется каталог `my_packet` со следующим содержимым:

```python
my_packet/
    __init__.py        # переменная __all__ = ['pack1', 'pack2']
    pack1/
        __init__.py    # переменная __all__ = ['module1', 'module2']
        module1.py
        module2.py
        module3.py
    pack2/             # переменная __all__ = ['module1', 'module2']
        __init__.py
        module1.py
        module2.py
    pack3/             # каталог не содержит файла __init__.py
        module1.py
        module2.py
```
Каждый файл `__init__.py` имеет вид:
```python
__all__ = ['pack1', 'pack2']
print(__name__)
```
Каждый файл `module#.py` имеет вид:
```python
def hello():
  return(__name__)

print(__name__)
```
В начале файла `my_packet/pack3/module2.py` присутствует строка `__all__ = []`.

Требуется рассмотреть различные варианты импорта содержимого пакета `my_packet`.

````python
from my_packet.pack1.module1 import hello # импорт функции hello из my_packet.pack1.module1
hello()
````

````python
import my_packet.pack1.module3 # импорт модуля из пакета my_packet.pack1, не указанного в __all__
my_packet.pack1.module3.hello()
````

````python
from my_packet import * # импорт всего пакета
````

````python
from my_packet.pack2 import * # импорт всех указанных в __all__ модулей, входящих в пакет my_packet.pack2
````

````python
from my_packet.pack3 import * # попытка импорта всех модулей, входящих в пакет my_packet.pack3 (ничего не происходит)
````

### Импорт по абсолютному и относительному пути

В Python 3.0 по умолчанию поиск импортируемого модуля производится в каталогах, перечисленных в списке `sys.path`. Этот способ поиска модулей называется *поиском по абсолютному пути*. Второй вариант поиска - *поиск по относительному пути* - подразумевает использование в инструкции `from` символа точки `.` и означает, что поиск импортируемого модуля необходимо вести в пределах пакета, содержащего текущий импортирующий модуль. Таким образом, при поиске по относительному пути модули внутри пакета получают преимущество перед модулями за его предалами.  

Введение двух способов импорта модулей обусловлено тем, что в ранних версиях Python поиск модулей всегда начинался с каталога пакета, в результате чего не существовало простого способа импорта модуля, например, из стандартной библиотеки, если его имя совпадало с именем модуля текущего пакета. В качестве решения данной проблемы предлагалось не использовать имена стандартной библиотеки.

**(!)** Символ `.` допустим только в команде `from`, в команде `import` его использовать нельзя.

**(!)** Резюме: команда `import` всегда осуществляет поиск по абсолютному пути, команда `from` может использовать абсолютный и относительный варианты поиска.

**Пример**  
Пусть в структуру каталога `my_packet` внесены следующие изменения:

```python
my_packet/
    __init__.py        
    pack1/
        __init__.py    
        module1.py
        module2.py
        module3.py
    pack2/             
        __init__.py
        module1.py
        module2.py
        module3.py  # содержит строку from .module1 import hello       
    pack3/
        module1.py
        module2.py
        module3.py  # содержит строку from ..pack1.module1 import hello
```
По-прежнему в каждом файле `module#.py` присутствует строка `print(__name__)`.

Требуется рассмотреть варианты импорта функции `hello` из модулей `module3.py`, расположенных в каталогах `pack2` и `pack3`.

````python
from my_packet.pack2.module3 import hello
hello()
````

````python
from my_packet.pack3.module3 import hello
hello()
````

### Общие сведения
*Пространства имён* - это области, в которых интерепретор Python хранит информацию об именах объектов и в которых он осуществляет поиск значения объекта по его имени.

Существует три пространства имён:
* [локальное](#Локальное-пространство-имён),
* [глобальное](#Глобальное-пространство-имён),
* [пространство встроенных имён](#Пространство-встроенных-имён) `__builtins__`.

Объект называется *локальным* для функции `fun`, если присваивание этому объекту происходит внутри блока определения функции `def`. Объект называется *нелокальным* для функции, если присваивание ему значения происходит внутри объемлющего функцию блока `def`. Переменная называется *глобальной*, если присваивание ей производится вне описания всех функций.  

Рассмотрим следующий пример:

```python
x = 1
...
def my_fun1():
    y = 2
    ...
    def my_fun2():
        z = 3
        ...
    ...

```

Здесь:
* переменная `x` - является *глобальной* переменной модуля;
* переменная `y` - является *локальной* для функции `my_fun1()` и *нелокальной* для функции `my_fun2()`;
* переменная `z` - является *локальной* для функции `my_fun2()`.

*Областью видимости переменной* называется область программы, где эта переменная может быть использована. 

Для рассмотренного примера областями видимости переменных `x`, `y` и `z` будут:
* весь модуль для перпеменной `x`;
* тело функций `my_fun1()` и `my_fun2()` для переменной `y`;
* тело функции `my_fun2()` для переменной `z`.

Области видимости устраняют конфликты имён, позволяя использовать переменные с совпадающими именами в разных областях вивдимости.

````python
x = 0

def my_fun():
    x = 1
    print(x)

my_fun()
print(x)  # вызов функции не привёл к изменению глобальной переменной x
````

### Разрешение имён

Разрешение имён в Python осуществляется в соответствии правилом *LEGB*, название которого образовано из названий областей видимости:
* L (Local) - локальная область видимости,
* E (Enclosing) - объемлющея оператор `def` область видимости,
* G (Global) - глобальная область видимости,
* B (Built-in) - встроенная область видимости.

Поиск завершается, как только будет найдено искомое имя. Если имя не нашлось ни в одной из перечсиленных областей Python генерирует сообщение об ошибке.

Полезным будет также помнить, что:
* глобальная область видимости относится к файлу модуля, т.е. не существует одной всеобемлющей области видимости;
* каждый вызов функции создаёт новую локальную область видимости;
* операция присваивания, выполняемая в теле функции, создаёт локальное имя, только если оно не было объявлено *глобальным* или *нелокальным* с помощью ключевых слов `global` и `nonlocal` соответственно.

Инструкция `global` сообщает Python, что функция будет изменять глобальную переменную, определённую на уровне модуля. Если требуется только получить значение глобальной переменной, то инструкция `global` не требуется.

Инструкция `nonlocal` сообщает Python, что функция будет изменять переменную, определённую на одном из объемлющих уровней. Как и с инструкцией `global`, для получения значения нелокальной переменной инструкция `nonlocal` не требуется.

**(!)** Использоание глобальных переменных существенно усложняет программу и повышает вероятность возникновения трудно обнаруживаемых ошибок.

**(!)** Отличие инструкций `global` и `nonlocal` заключается в следующем:
* инструкция `global` вынуждает Python начинать поиск с глобальной области видимости и продлжать при необходимости во встроенной области, в то время как инструкция `nonlocal` предписывает начать поиск с ближайшей объемлющей области поиска вплоть до глобальной области видимости, не включая последнюю и игнорируя встроенную область (т.е. в обоих слачаях имеем отклонение от правила LEGB);
* инструкция `nonlocal` требует, чтобы к моменту её выполнения объект с указанным именем уже существовал, у инструкции `global` - такого требования нет.

````python
x = 0  # глобальная переменная

def my_fun():
    print(x)  # для доступа к значению глобальной переменной инструкция global не требуется
    
my_fun()
print(x)
````

````python
x = 0  # глобальная переменная

def my_fun():
    x = 1    # присваивание значения локальной переменной 
             # не изменяет значения одноимённой глобальной переменной
    print(x)
    
my_fun()
print(x) 
````

````python
x = 0  # глобальная переменная

def my_fun():
    global x  # переменная x связывается с глобальной переменной x
    x = 1     # присваивание значения глобальной переменной 
    print(x) 
    
my_fun()
print(x) 
````

````python
def my_fun1():
    global x
    x = 1
    
def my_fun2():
    global x
    x = 2
    
def asd(a,b):
    c = a+b
    return c

d = asd(10,20)
print(d)
````

````python
# Порядок вызова влияет на результат вывода.
my_fun1(); print(x)
my_fun2(); print(x)
````

````python
# Порядок вызова влияет на результат вывода.
my_fun2(); print(x)
my_fun1(); print(x)
````

````python
x = 0

def my_fun1():
    x = 1
    def my_fun2():
        x = 2
        print(x)

    my_fun2()
    print(x)

print(x)
my_fun1()
print(x)
````

### Замыкания

*Замыканием* или *фабричной функцией* называется объект-функция, которая сохраняет своё состояние в объемлющих областях видимости. Достигается это за счёт использования нелокальных переменных, определённых в объемлющих областях видимости. Самым неочевидным здесь является то, что значение переменной, хранящей состояние, продолжает существовать даже после уничтожения объемлющей области видимости.

**(!)** Если вложенная функция ссылается на переменную цикла из объемлющей области видимости, то сохранено будет последнее значение переменной цикла.

````python
def my_fun(power):
    p = power
    def power_fun(x):  # создание функции, поведение которой 
        return(x**p)   # зависит от значения нелокальной переменной p
    
    return(power_fun)  # созданная функция возвращается как результат

power1 = my_fun(1)  # создание функции
print(power1(2))    # применение

power2 = my_fun(2)  # создание функции
print(power2(2))    # применение

power3 = my_fun(3)  # создание функции
print(power3(2))    # применение
````

## Элементы объектно-ориентированного программирования

*Классы* в Python позволяют программисту определять свои собственные типы данных с их *свойствами* (*атрибутами*) и *методами*.

Классы подобно модулям определяют свои пространства имён. Каждый раз, когда выполняется инструкция `class` создаётся *объект класса*, а когда вызывается *конструктор класса*, создаётся новый объект - *экземпляр класса* со своим собственным пространством имён, отличающимся от других экземпляров. Можно сказать, что классы выполняют роль фабрик по производству экземпляров классов (в языках программирования C++, Pascal, Java используется термин *объект*, но в Python всё является объектами, в том числе сами классы).

Классы Python поддерживают *множественное наследование* - механизм адаптации поведения класса. Экземпляры класса наследуют атрибуты и методы порождающих их классов, которые, в свою очередь, наследуют атрибуты и методы породивших их классов и т.д. вверх по иерархии до базового класса *object*, являющегося родителем всех классов. Класс, который используется для порождения нового класса, называется *базовым*, *родительским* или *суперклассом*, класс, который порождается - *дочерним* или *подклассом*.

Существует специальная группа *магических методов* (функций, оформление которых выполняется в соответствии со специальным протоколом), позволяющих переопределять операции, выполняемые над экземплярами классов. Применение магических методов может сделать использование экземпляров класса более естественным (похожим на использование базовых типов).

В идеале программирование в стиле ООП сводится к сборке и отладке из уже отлаженных блоков новых компонентов   путём написания собственных подклассов. Естественно такой идеал может быть достигнут либо путём использования уже готовых библиотек классов либо тщательным предварительным проектированием своих собственных библиотек.

### Создание классов

Для создания класса предназначена инструкция `class`.

`class` - это составная инструкция, которая может содержать блок операторов, включающий в себя все допустимые операции Python (`=`, `def`, `if`, `for`, `while`,...). 

**(!)** Важно понимать, что `class` в отличие от, например, C++ - это не объявление класса, а самая настоящая инструкция, которая должны быть выполнена, чтобы классом можно было пользоваться. При выполнении инструкции `class` выполняются все содержащиеся в ней команды.

Cразу после ключевого слова `class` должно следовать имя создаваемого класса. Имена класов по принятому соглашению  должны начинаться с прописной (большой) буквы. После имени класса в скобках через запятую могут быть перечислены имена родительских *суперклассов*. Если суперклассов нет, то создаваемый класс будет наследоваться от класса `object`, скобки в этом можно опустить. Заканчиваться инструкция `class` должна символом двоеточия. 

После двоеточия следует блок операторов. Любая операция присваивания `=` в этом блоке создаёт *атрибут класса* (аналог статических членов в C++), любая операция объявления функции `def` создаёт *метод класса*.

Результатом выполнения инструкции `class` является *объект класса*, *вызов объекта класса* приводит к созданию *экземпляра класса*.

Присваивание *атрибуту экземпляра класса* приводит к созданию/изменению атрибута в самом экземпляре.

Атрибуты класса доступны всем экземплярам класса, атрибуты экземпляров классов доступны только им самим.

У всех методов класса первым списке обязательно должен аргумент, с помощью которого Python передаёт методу ссылку на экземпляр класса. Принято этому аргументу давать имя `self` (аналог `this` в С++). Таким образом, все методы класса имеют как минимум один аргумент. 

Вызов метода экземпляра класса `obj.method(...)` автоматически транслируется в вызов метода класса `class.method(obj,...)`. При желании можно пользоваться любым способом.

Метод `__init()__` является *конструктором класса* и вызвается всякий раз, когда происходит вызов объекта класса с целью создания экземпляра класса.

### Наследование

*Наследование* по сути сводится к операции поиска атрибутов/методов в дереве пространств имён. Каждый раз, когда интерпретатор Python встречает выражение `object.attr` или `object.method()` (где `object` - это либо экземпляр класса, либо объект класса) запускается процедура поиска первого вхождения атрибута/метода в дереве пространств имён в направлении снизу вверх слева направо, начиная с объекта `object`.

Обычно классы, порождаемые от других классов, изменяют поведение родительских классов, переопределяя их методы. При этом поведение новых классов может кардинально отличаться от поведения родительских классов. Но часто бывают ситуации, когда необходимо внести изменение в поведение родительского класса, не нарушив кардинально логику его работы. Обычно вызов методов осуществляется через обращение к экземпляру класса `object.method()`, но в данной ситуации оказывается полезным вызов метода посредством экземпляра класса `class.method()`, позволяющий осуществить вызов метода родительского класса.

Особенно часто потребность в вызове методов родительского класса возникает в конструкторе класса. Например, при создании нового экземпляра класса `My_class`, унаследованного от класса `My_super_class`, может потребоваться вначале произвести базовую настройку экземпляра и только потом произвести его донастройку. В следующем примере при создании экземпляра класса `My_class` вначале будет осуществлён вызов конструктора родительского суперкласса, который получит ссылку на создаваемый экземпляр и произведёт его базовую настройку, и только потом будет произведена донастройка создаваемого экземпляра. 

```python
class My_super_class():
    def __init__(self, ...):  # конструктор родительского суперкласса, выполняющий базовую настройку
        ...
        
class My_class():
    def __init__(self, ...):
        My_super_class.__init__(self, ...)  # вызов конструктора родительского суперкласса для базовой настройки 
        ...                                 # дальнейшая настройка создаваемого экземпляра
```
 

````python
class My_super_class1:
    def method(self):
        print('My_super_class1: method')
    
    def method1(self):
        print('My_super_class1: method1')

class My_super_class2:
    def method(self):
        print('My_super_class2: method')
    
    def method1(self):
        print('My_super_class2: method1')

    def method2(self):
        print('My_super_class2: method2')
        
class My_class(My_super_class1, My_super_class2):
    def method(self):
        print('My_class: method')
````

````python
o = My_class()

o.method()
o.method1()
o.method2()
````

````python
class My_super_class():
    def __init__(self, arg):
        self.attr = arg
        
class My_class(My_super_class):
    def __init__(self, arg1, arg2):
        My_super_class.__init__(self, arg1)
        self.attr = self.attr ** arg2
````

````python
o = My_class(2, 8)
o.attr
````

````python
# Практический пример: класс вектор на плоскости.

class My_point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def to_string(self):
        return f'x = {self.x}, y = {self.y}'
    
class My_vector(My_point):
    def __init__(self, x, y):
        My_point.__init__(self, x, y)
        
    def add(self, v):
        return My_vector(self.x + v.x, self.y + v.y)

    def scalar(self, n):
        return My_vector(n * self.x, n * self.y)
````

````python
v1 = My_vector(1, 0)
v2 = My_vector(0, 1)
print('v1: ', v1.to_string())
print('v2: ', v2.to_string())

v3 = v1.add(v2)
print('v3: ', v3.to_string())

v4 = v3.scalar(2)
print('v4: ', v4.to_string())
````

### Магические методы

*Магическими методами* принято называть методы, которые обрамлены с двух сторон двумя символами подчёркивания, например, `__init__()` и которые используются интерпретатором Python для выполнения отображения операторов выражений на эти методы (и не только). 

С подробным описанием магических методов можно ознакомиться на странице [официальной документации](https://docs.python.org/3/contents.html), в разеделе [Special method names](https://docs.python.org/3/reference/datamodel.html?highlight=__add__#special-method-names)

Так как список магических методов достаточно длинный, то рассмотрим с целью ознакомления только с некоторые из них.

* `__add__(self, other)` - эквивалентно выражению `object + other`;
* `__radd__(self, other)` - эквивалентно выражению `other + object`;
* `__sub__(self, other)` - эквивалентно выражению `object - other`;
* `__rsub__(self, other)` - эквивалентно выражению `other - object`;
* `__mul__(self, other)` - эквивалентно выражению `object * other`;
* `__rmul__(self, other)` - эквивалентно выражению `other * object`;
* `__str__(self)` - эквивалентно выражению `print(object)`;
* `__repr__(self)` - вызывается, когда интерпретатору требуется вывести информацию об объекте `object`.

````python
# Практический пример: класс вектор на плоскости.

class My_point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f'x = {self.x}, y = {self.y}'

    def __repr__(self):
        return f'x = {self.x}, y = {self.y}'    
````

````python
p = My_point(1,1)
print(p)
p
````

````python
class My_vector(My_point):
    def __init__(self, x, y):
        My_point.__init__(self, x, y)
        
    def __str__(self):
        return f'vector {My_point.__str__(self)}'
    
    def __repr__(self):
        return f'vector {My_point.__str__(self)}'
    
    def __add__(self, other):
        return My_vector(self.x + other.x, self.y + other.y)
    
    def __radd__(self, other):
        return My_vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return My_vector(self.x - other.x, self.y - other.y)
    
    def __rsub__(self, other):
        return My_vector(self.x - other.x, self.y - other.y)

    def __mul__(self, other):
        if isinstance(other, float) or isinstance(other, int):
            return My_vector(other * self.x, other * self.y)
        else:
            return self.x * other.x + self.y * other.x

    def __rmul__(self, other):
        if isinstance(other, float) or isinstance(other, int):
            return My_vector(other * self.x, other * self.y)
        else:
            return self.x * other.x + self.y * other.x
````

````python
v1 = My_vector(1,0)
v2 = My_vector(0,1)
````

````python
v1 + v2
````

````python
v1 - v2
````

````python
2*(v1 + v2)
````

````python
v1 * v2
````

### Работа с текстовыми файлами
Для работы с файлами (чтение, запись) предназначена функция `open`. Синтаксис функции следующий:

```python
f = open(filename, mode)
```
Функция `open` возвращает объект, который связан с файлом `filename` открытым в режиме `mode`.


Возможны следующие значения аргумента `mode`:  
* `r` - файл открывается на чтение,  
* `w` - запись в файл (при этом существующий файл будет полностью перезаписан),  
* `a` - добавление в конец файла,  
* `r+` - чтение и запись.  

**(!)** Все перечисленные режимы являюся *текстовыми*.  Для работы с *двоичными* данными  предназначены режимы: `br`, `bw`, `ba`, `br+`.

Файловый объект поддержимвает следующие методы:
* `read([size])` - читает из файла `size` байт, если `size` опущен, считывается весь файл;  
* `readline()` - считывает из файла одну строку;  
* `readlines()` - возвращает список строк файла;
* `write(string)` - записывает в файл значение строки `string`;  
* `writelines(lines)` - записывает в файл список строк объекта `lines` (символы перевода строк автоматически не добавляются);
* `tell()` - возвращает указатель текущей позиции;   
* `seek(offset, [from_what])` - смещает указатель текущей позиции в файле на величину `offset` относительно положения `from_what` (`0` - от начала файла (значение по умолчанию), `1` - от текущего положения, `2` - от конца файла);
* `flush()` - сбрасывает (выталкивает) содержимое файлового буфера на диск;
* `close()` - закрывает файл.

**(!)** Метод `readline()` всегда возвращает строку, заканчивающуюся символом перевода строки `\n`,  
и только если была прочитана последняя строка или был достигнут конец файла, будет возвращена строка без символа перевода строки.

**(!)** Python всегда интерпретирует содержимое файлов как строки. Все операции чтения и записи оперируют строками.  Работа по преобразованию данных к требуемому типу ложится на плечи программиста.

**(!)** Метод `close()` не является обязательным - файл будет автоматически закрыт, как только исчезнет последняя ссылка на объект файла. Тем не менее, такое поведение Python не гарантируется в будущем и может быть изменено, поэтому явный вызов метода `close()` является хорошим стилем. Кроме того, закрытие файла всегда приводит к сбросу на диск содержимого буфера. Не учёт этого фактора, может сильно усложнить отладку программ.


### Применение менеджера контекста и итераторов для работы с файлами

Файловый объект относится к числу *итерируемых объектов*, говорят *"поддерживает протокол итераций"*. Проще говоря, файловый объект "умеет" возвращать свое содержимое последовательно в ответ на запрос "дай следующий элемент". Формально, каждый раз, когда осуществляется вызов метода `f.readline()` файлового объекта `f`, автоматически  происходит переход к следующей строке файла. 

Цикл `for` "умеет" работать с объектами, поддерживающими протокол итератора. Применение цикла `for` для чтения содержимого файла делает код компактнее. Например, следующий код

```python
f = open('filename')

line = f.readline()
while line != '':
    do_something()
    line = f.readline()

f.close()
```

может быть переписан так

```python
for line in open('filename'):
    do_something(line)
```

В последнем фрагменте будет создан временный файловый объект и выполенно его построчное чтение. Когда цикл закончит работу, временный файловый объект будет удалён, а связанный с ним файл будет автоматически закрыт.

*Менеджер контекста* позволяет обернуть программный код работы с содержимым файла дополнительным слоем логики, гарантирующим уничтожение файлового объекта, влекущее закрытие файла, после выхода за пределы менеджера.

```python
with open('filename') as my_file:
    for line in my_file:
        do_something(line)
```

````python
with open('test.txt', mode='w') as my_file:
    for i in range(1,4):
        my_file.writelines('line %s \n' % str(i))
        
with open('test.txt') as my_file:
    for line in my_file:
        print(line, end='')
````

### Модуль pickle

Модуль `pickle` позволяет сохранить в файле любые объекты Python.

Два основных метода, для сохранения и чтения объектов модуля `pickle` - это:
* `dump(object, file_object)` - сохраняет объект `object` в файле;
* `load(object, file_object)` - читает объект `object` из файла.

````python
# Сохранение в файле произвольных объектов с помощью pickle.

import pickle

class My_point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __eq__(self, other):
        return (self.x == other.x) and (self.y == other.y)
        
p1 = My_point(0, 0)
p2 = My_point(1, 1)

with open('test.txt', mode='bw') as f:
    pickle.dump(p1, f)
    pickle.dump(p2, f)
````

````python
# Просмотр содержимого файла с сохранёнными объектами pickle.

with open('test.txt', mode='br') as f:
    for line in f:
        print(line)
````

````python
# Восстановление данных из файла с помощью pickle.

with open('test.txt', mode='br') as f:
    p3 = pickle.load(f)
    p4 = pickle.load(f)

print(f'p1 = p3 is {p1 == p3}')
print(f'p2 = p4 is {p2 == p4}')
````

## Обработка исключительных ситуаций

Ошибки, возникающие в процессе работы программы и делающие невозможным её дальнейшее выполнение, называются *исключениями* (*exception*). При возникновении исключения выполнение программы прекращается и на экран выводится сообщение об ошибке с указанием информации о месте и характере ошибки.

Например, попытка вычислить следующее выражение
```python
x = 1/0
```

приведёт к возникновению исключительной ситуации и на экране появится информация следующего содержания:
```
---------------------------------------------------------------------------
ZeroDivisionError                         Traceback (most recent call last)
<ipython-input-76-bfdf2b0f658a> in <module>()
----> 1 x = 1/0

ZeroDivisionError: division by zero
```

Последняя строка содержит описание исключительной ситуации `ZeroDivisionError`, означающее попытку деления на 0.

Исключительные ситуации можно *перехватывать* с целью выполнения определённых действий, связанных с возникшей ошибкой или чтобы просто их игнорировать. Для этих целей в языке Python предназначена инструкция `try` - `except` - `else` - `finally`. Синтаксис инструкции следующий:

```python

try:
    # блок, в котором может произойти исключение

except Exception1:
    # блок, отвечающий за обработку исключения Exception1
    
except Exception2,  Exception3:
    # блок, отвечающий за обработку любого из исключений Exception2 или  Exception3

else:
    # блок, который выполняется если исключений не возникло
    
finally:
    # блок, отвечающий за выполнение завершающих операций, 
    # которые необходимо выполнить независимо от того, 
    # произошло ислючение или нет
   
```

````python
# Задача: сохранить в файле два числа, потом загрузить их из файла и вернуть их отношение.

def write_data(filename, value1, value2):  # функция сохраняет в файле с именем filename два значения
    with open(filename, mode='w') as f:
        f.writelines(str(value1) + '\n')
        f.writelines(str(value2) + '\n')

def read_data(filename): # функция читает из файла с именем filename два значения и возвращает их отношение
    f = open(filename, mode='r') # опасное место - файла, с таким именем может не существовать и т.п.
    x = float(f.readline())      # опасное место - преобразование строки в число может закончиться ошибкой
    y = float(f.readline())      # опасное место - преобразование строки в число может закончиться ошибкой
    f.close()                    # опасное место - из-за ошибки, произошедшей раньше, файл останется открытым
    return x / y                 # опасное место - может быть деление на ноль
````

````python
# Решение задачи состоит из трёх попыток, две из которых приводят к возникновению ошибки.

fname = 'temp.txt'  # имя временного файла
n = 0  # счётчик попыток, их будет 3

while True:
    try:
        n = n + 1  # увеличиваем счётчик попыток на 1 и выводим его на экран
        print('Attempt', n)

        if n == 1:  # первая попытка - сохраняем в файле вместо чисел строки
            write_data(fname, 'hello', 'world')  
            result = read_data(fname)

        if n == 2:  # вторая попытка - сохраняем два числа, но знаменатель равен 0
            write_data(fname, 1, 0)
            result = read_data(fname)

        if n == 3: # третья попытка - делаем всё правильно, сохраняем два числа 
            write_data(fname, 10, 2)
            result = read_data(fname)

    except ValueError:  # обрабатываем исключение, возникшее при попытке преобразовать строки в числа
        print('Wrong values')

    except ZeroDivisionError:  # обрабатываем исключение из-за попытки деления на ноль
        print('Division by zero')

    else:  # задача решена, ошибок в процессе решения не возникло 
        print('result =', result)
        break  # только решив задачу без ошибок мы сможем покинуть вечный цикл While True

    finally:
        import os
        os.remove('temp.txt')  # неважно чем закончилось решение задачи, необходимо удалить временный файл
        print('This is the end of attempt.\n')
        
print('This is the end.')
````
